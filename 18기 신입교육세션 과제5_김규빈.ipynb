```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
import warnings
warnings.filterwarnings('ignore')
```

```python
# train ë°ì´í„°, test ë°ì´í„° ë¶„í• 
X_train = data_train.drop(columns=['Churn']).copy()
Y_train = data_train['Churn'].copy()

X_test = data_test.drop(columns=['Churn']).copy()
Y_test = data_test['Churn'].copy()
```

```python
# ë²”ì£¼í˜• ë³€ìˆ˜ One-hot ì¸ì½”ë”© + train ê¸°ì¤€ìœ¼ë¡œ test ì»¬ëŸ¼ ë§ì¶”ê¸°
cols_ohe = ['Gender', 'Subscription Type', 'Contract Length']

X_train = pd.get_dummies(X_train, columns=cols_ohe, drop_first=True)
X_test = pd.get_dummies(X_test, columns=cols_ohe, drop_first=True)

X_test = X_test.reindex(columns=X_train.columns, fill_value=0)
```

```python
# train ë°ì´í„°, valid ë°ì´í„° ë¶„í• 
X_tr, X_val, Y_tr, Y_val = train_test_split(
    X_train, Y_train, test_size=0.2, random_state=42, stratify=Y_train
)
```

```python
# í‰ê°€ í•¨ìˆ˜
def eval_classification(model, X_tr, X_val, Y_tr, Y_val):
    
    # ì˜ˆì¸¡ê°’
    Y_tr_pred = model.predict(X_tr)
    Y_val_pred = model.predict(X_val)
    
    # í™•ë¥ ê°’
    Y_tr_proba = model.predict_proba(X_tr)[:,1]
    Y_val_proba = model.predict_proba(X_val)[:,1]
    
    print('í•™ìŠµ ëŠ¥ë ¥')
    print("Accuracy:", accuracy_score(Y_tr, Y_tr_pred))
    print("F1-score:", f1_score(Y_tr, Y_tr_pred))
    print("ROC-AUC:", roc_auc_score(Y_tr, Y_tr_proba))
    
    print('ì¼ë°˜í™” ëŠ¥ë ¥')
    print("Accuracy:", accuracy_score(Y_val, Y_val_pred))
    print("F1-score:", f1_score(Y_val, Y_val_pred))
    print("ROC-AUC:", roc_auc_score(Y_val, Y_val_proba))
```

```python
# í•™ìŠµ ìˆ˜í–‰
lr = LogisticRegression(
    max_iter=1000,
    random_state=42,
    n_jobs=-1
)

dt = DecisionTreeClassifier(
    max_depth=5,
    random_state=42
)

rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=12,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)

gbr = GradientBoostingClassifier(
    n_estimators=150,
    learning_rate=0.05,
    max_depth=3,       
    subsample=0.8,
    random_state=42
)

lr.fit(X_tr, Y_tr)
dt.fit(X_tr, Y_tr)
rf.fit(X_tr, Y_tr)
gbr.fit(X_tr, Y_tr)
```

```python
# Logistic Regression ì„±ëŠ¥
eval_classification(lr, X_tr, X_val, Y_tr, Y_val)
```

```python
# Decision Tree ì„±ëŠ¥ 
eval_classification(dt, X_tr, X_val, Y_tr, Y_val)
```

```python
# Random Forest ì„±ëŠ¥
eval_classification(rf, X_tr, X_val, Y_tr, Y_val)
```

```python
# Gradient Boosting ì„±ëŠ¥
eval_classification(gbr, X_tr, X_val, Y_tr, Y_val)
```

```python
from sklearn.pipeline import make_pipeline # íŠ¹ì„±ê³µí•™ + í•™ìŠµ
from sklearn.impute import SimpleImputer # ê²°ì¸¡ê°’ ë‹¨ìˆœ ëŒ€ì¹˜ ê¸°ë²•
from sklearn.ensemble import GradientBoostingClassifier # í•™ìŠµ ì•Œê³ ë¦¬ì¦˜
from sklearn.preprocessing import StandardScaler #Standard Scaling
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform
```
