# **과제**
**1. 추가적으로 EDA 진행해서 '생존'과 연관된 변수 찾기 + insight 정리**
   - 가설 검정, 통계분석 이용하기

### **1.1 성별(Sex)과 생존 여부**
- "여성이 우선적으로 구조되었다"는 도메인 지식 존재
- 성별과 생존 여부 간의 관계를 교차표, 시각화, 카이제곱 검정을 통해 확인

```python
# 성별과 생존 여부의 교차표
ct_sex = pd.crosstab(data["Sex"], data["Survived"])
ct_sex
```

```python
# 히트맵 시각화
sns.heatmap(ct_sex, annot=True, fmt="d", cbar=False)
plt.title("2x2 Contingency Table (Sex x Survived)")
plt.xlabel("Survived (0=Died, 1=Survived)")
plt.ylabel("Sex")
```

```python
# 성별과 생존 여부의 독립성 검정
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(ct_sex)

print(f"Chi-square statistic: {chi2:.4f}")
print(f"p-value: {p:.4f}")
print(f"Degrees of freedom: {dof}")
```

- 귀무가설(H0): 성별과 생존 여부는 독립이다.
- p-value < 0.05이므로 귀무가설을 기각한다.

    즉, 성별과 생존 여부는 통계적으로 유의한 관계가 있으며,
    이는 도메인 지식과도 일치하는 결과이다.

### **1.2 객실 등급(Pclass)과 생존 여부**

```python
# 객실 등급과 생존 여부 교차표
ct_pclass = pd.crosstab(data1["Pclass"], data1["Survived"])

sns.heatmap(ct_pclass, annot=True, fmt="d", cbar=False)
plt.title("2x2 Contingency Table (Pclass x Survived)")
plt.xlabel("Survived (0=Died, 1=Survived)")
plt.ylabel("Pclass")
```

```python
# 카이제곱 검정
chi2, p, dof, expected = chi2_contingency(ct_pclass)

print(f"Chi-square statistic: {chi2:.4f}")
print(f"p-value: {p:.4f}")
print(f"Degrees of freedom: {dof}")
```

객실 등급과 생존 여부는 통계적으로 유의한 관계가 있다. <br>
1등석 승객의 생존률이 가장 높고, 3등석 승객의 생존률이 가장 낮게 나타난다.

### **1.3 연령(Age) 구간화**
- Child(0-12세) : 보호가 필요한 아동으로, 구조 우선 대상일 가능성 높음
- Teen(13-18세) : 아동과 성인의 중간 단계
- Adult(19-60세) : 성인 승객의 대다수를 차지하는 연령대
- Senior(60세 이상) : 신체적 제약으로 인해 구조에 취약할 가능성이 있는 집단

```python
# 연령을 사회적 역할 및 구조 우선순위를 고려한 구간으로 분할
data1["AgeGroup"] = pd.cut(
    data1["Age"],
    bins=[0, 12, 18, 60, 100],
    labels=["Child", "Teen", "Adult", "Senior"]
)

pd.crosstab(data1["AgeGroup"], data1["Survived"], normalize="index")
```

### **1.4 AgeGroup과 성별(Sex)의 결합 분석**

```python
ct_age_sex = pd.crosstab(
    [data1["AgeGroup"], data1["Sex"]],
    data1["Survived"]
)

# 생존률로 변환
survival_rate = ct_age_sex.div(ct_age_sex.sum(axis=1), axis=0)
survival_rate
```

```python
sns.heatmap(survival_rate, annot=True, fmt=".2f", cmap="Blues")
plt.title("Survival Rate by AgeGroup and Sex")
```

- 아동 그룹에서는 성별 간 생존률 차이가 상대적으로 작다.
- 성인 및 고령층에서는 여성의 생존률이 남성보다 현저히 높다.

    이는 구조 과정에서 연령과 성별이 동시에 고려되었음을 시사한다.

### **2.1 Title 변수 생성**

```python
# Name 구조 확인
data2["Name"].head()
```

```python
# data2["Title"] = data2["Name"].str.extract(r",\s*([^\.]+)\.")
data2["Title"] = data2["Name"].apply(lambda x: x.split(",")[1].split(".")[0].strip())
data2[["Name", "Title"]].head()
```

### **2.2 희귀 Title 통합**

- Mr: 18세 이상 남성
- Miss(Mlle) : 미혼 여성
- Mrs(Mme) : 기혼 여성
- Master : 13세이하 남성
- Dr : 박사, 의사
- Rev: 성직자
- Col, Major, Capt: 군 장교
- Lady, the Countess: 귀족 여성
- Don, Sir, Jonkheer : 귀족 남성

- Dr, Rev, Col, Major, Capt, Sir, Don, Jonkheer -> Mr
- Lady, the Countess, Mme -> Mrs
- Mlle, Ms -> Miss

```python
data2["Title"].value_counts()
```

```python
# 희귀 title 성별 분포 확인
rare_candidates = data2[~data2["Title"].isin(["Mr", "Miss", "Mrs", "Master"])]
pd.crosstab(
    rare_candidates["Title"],
    rare_candidates["Sex"]
)
```

```python
# 희귀 title 나이 분포 확인
rare_candidates.groupby("Title")["Age"].describe()
```

```python
title_mapping = {
    # Mr 계열 (성인 남성)
    "Dr": "Mr",
    "Rev": "Mr",
    "Col": "Mr",
    "Major": "Mr",
    "Capt": "Mr",
    "Sir": "Mr",
    "Don": "Mr",
    "Jonkheer": "Mr",

    # Mrs 계열 (성인 여성)
    "Lady": "Mrs",
    "the Countess": "Mrs",
    "Mme": "Mrs",

    # Miss 계열 (미혼 여성)
    "Mlle": "Miss",
    "Ms": "Miss"
}

data2["Title"] = data2["Title"].replace(title_mapping)
data2["Title"].value_counts()
```

### **2.3 Age 결측값 채우기**
- Title + Pclass 기준

```python
# Age 결측치 개수 확인
data2["Age"].isnull().sum()
```

```python
# Title + Pclass 조합별 표본 수 확인
data2.groupby(["Title", "Pclass"])["Age"].count()
```

```python
# Title별 Pclass 중앙값 계산
age_median = (data2.groupby(["Title", "Pclass"])["Age"].median())
age_median
```

```python
# Age 결측치 대체
data2["Age"] = data2["Age"].fillna(
    data2.groupby(["Title", "Pclass"])["Age"].transform("median")
)
```

```python
# 결측치 제거여부 확인
data2["Age"].isnull().sum()
```

- 데이터를 Ai에 입력한 뒤 분석을 맡기는 행위는 금지  
- **개인정보 또는 기업 내부 데이터를 ChatGPT에 입력하는 것은 절대 금지**  

2) 자료 조사·레퍼런스 확인(가능)  
- ChatGPT를 활용해 관련 자료/기사/논문/공식 문서의 링크와 출처를 요청 가능  
- 단, 반드시 원문을 직접 확인하고 필요한 내용은 검증 후 사용  

3) 코딩 지원(적극 활용 권장)  
- 코드 작성/디버깅/리팩토링 등은 적극적으로 활용  
- 단, ChatGPT가 제시한 코드는 그대로 복붙하지 말고, 원리를 이해한 뒤 실행하고 결과를 검증
