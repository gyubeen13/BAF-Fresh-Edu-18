  - 출처 : kaggle(https://www.kaggle.com/datasets/juhibhojani/house-price)
  - 분야: 금융(부동산)
  - 내용 : 총 1,295채 주택의 판매 데이터를 기반으로, 면적, 화장실 개수, 층수, 주차장 유무, 거리 등 13개 특성을 활용하여 최종 주태개 판매가격을 예측할 수 있는 데이터입니다.

  - 주택 가격 결정 요인 분석을 통해 부동산 가치 평가, 금융권 담보 평가, 시장 가격 예측 등에 활용할 수 있으며, 비교적 결측치가 적고 변수 구조가 명확하여 데이터 부석 및 모델링 학습용으로 적합할 것 같습니다.

  - 출처 : 서울열린데이터광장(https://data.seoul.go.kr/dataList/OA-15572/S/1/datasetView.do)
